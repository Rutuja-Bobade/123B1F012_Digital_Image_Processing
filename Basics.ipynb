```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img, title=""):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # BGR â†’ RGB for matplotlib
    plt.imshow(img_rgb)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
# Drawing a simple line
img = np.zeros((400,400,3),dtype = 'uint8')
cv2.line(img,(50,50),(350,350),(150,0,180),10)
cv2.line(img,(100,100),(300,300),(150,0,180),10)
show(img, "Lines")
```

```python
# Drawing a rectangle along with previous lines
cv2.rectangle(img, (100,100), (300,300), (100,0,100), 10)
show(img, "Rectangle + Lines")
```

```python
# Drawing a rectangle without previous image
img1 = np.zeros((400,400,3), dtype='uint8')
cv2.rectangle(img1, (100,100), (300,300), (100,0,180), 10)
show(img1, "Rectangle Only")
```

```python
# Drawing a hollow circle
img2 = np.zeros((400,400,3), dtype='uint8')
cv2.circle(img2, (200,200), 100, (100,0,100), 10)
show(img2, "Circle")
```

```python
#drawing a triangle
img3 = np.zeros((400,400,3),dtype = 'uint8')
```

```python
# Drawing a triangle
img3 = np.zeros((400,400,3), dtype='uint8')
p1 = (300,300)
p2 = (100,200)
p3 = (180,150)

cv2.line(img3, p1, p2, (100,50,100), 8)
cv2.line(img3, p2, p3, (100,50,100), 8)
cv2.line(img3, p3, p1, (100,50,100), 8)
show(img3, "Triangle")
```
