```python
imp_noise=np.zeros((640,480),dtype=np.uint8)
cv2.randu(imp_noise,0,255)
_, imp_noise=cv2.threshold(imp_noise,245,255,cv2.THRESH_BINARY)
imp_noise = np.array(imp_noise, dtype=np.uint8)
```

```python
imp_noise = cv2.resize(imp_noise, (image.shape[1], image.shape[0]))
```

```python
impulse_img = cv2.add(image, imp_noise)
```

```python
fig=plt.figure(dpi=300)

fig.add_subplot(1,3,1)
plt.imshow(image,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,3,2)
plt.imshow(imp_noise,cmap='gray')
plt.axis("off")
plt.title("Impulse Noise")

fig.add_subplot(1,3,3)
plt.imshow(impulse_img,cmap='gray')
plt.axis("off")
plt.title("Combined")
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(impulse_img,cmap='gray')
plt.title("with impulse noise")
plt.axis("off")

plt.subplot(1,2,2)
plt.hist(impulse_img.ravel() , bins=256 , range=[0,255], color='black')
plt.title("Histogram of impulse image")
plt.tight_layout()
plt.show()
```

```python
sp_noise = np.zeros((630, 1200), dtype=np.uint8)
cv2.randu(sp_noise, 0, 255)
salt_prob = 0.02
pepper_prob = 0.02

_, salt = cv2.threshold(sp_noise, 255 * (1 - salt_prob), 255, cv2.THRESH_BINARY)
_, pepper = cv2.threshold(sp_noise, 255 * pepper_prob, 255, cv2.THRESH_BINARY_INV)

sp_noise = cv2.add(salt, pepper)
```

```python
salt_pepper_img = cv2.add(image, sp_noise)
```

```python
fig=plt.figure(dpi=300)

fig.add_subplot(1,3,1)
plt.imshow(image,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,3,2)
plt.imshow(sp_noise,cmap='gray')
plt.axis("off")
plt.title("Impulse Noise")

fig.add_subplot(1,3,3)
plt.imshow(salt_pepper_img,cmap='gray')
plt.axis("off")
plt.title("Combined")
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(salt_pepper_img,cmap='gray')
plt.title("with impulse noise")
plt.axis("off")

plt.subplot(1,2,2)
plt.hist(salt_pepper_img.ravel() , bins=256 , range=[0,255], color='black')
plt.title("Histogram of impulse image")
plt.tight_layout()
plt.show()
```
