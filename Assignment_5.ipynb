```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('nature-images.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(gray, cmap="gray")
plt.title("Gray Image")
plt.axis("off")

plt.show()
```

```python
# Formula: new_pixel = alpha * pixel + beta
# alpha < 1 reduces contrast, beta shifts brightness
low_contrast = cv2.convertScaleAbs(gray, alpha=0.5, beta=30)
```

```python
plt.figure(figsize=(6, 6))

plt.imshow(low_contrast, cmap='gray')
plt.title("Simulated Low-Contrast Image")
plt.axis("off")
plt.show()
```

```python
min_val = np.min(low_contrast)
max_val = np.max(low_contrast)
contrast_stretched = ((low_contrast - min_val) / (max_val - min_val)) * 255
contrast_stretched = contrast_stretched.astype(np.uint8)
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis("off")
plt.show()
```

```python
hist_equalized = cv2.equalizeHist(low_contrast)

plt.figure(figsize=(6, 6))
plt.imshow(hist_equalized, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(low_contrast, cmap='gray')
plt.title("Low-Contrast Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(contrast_stretched, cmap='gray')
plt.title("After Contrast Stretching")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(hist_equalized, cmap='gray')
plt.title("After Histogram Equalization")
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(low_contrast.ravel(), bins=256, range=[0, 256], color='gray')
plt.title("Low-Contrast Image Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(1, 3, 2)
plt.hist(contrast_stretched.ravel(), bins=256, range=[0, 256], color='gray')
plt.title("Contrast Stretched Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(1, 3, 3)
plt.hist(hist_equalized.ravel(), bins=256, range=[0, 256], color='gray')
plt.title("Histogram Equalized Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
